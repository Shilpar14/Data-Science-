```python
group1_setosa=df[df['target']==0]['sepal width (cm)']
group2_versicolor=df[df['target']==1]['sepal width (cm)']
group3_virginica=df[df['target']==2]['sepal width (cm)']
```

```python
group1_setosa.mean(),group2_versicolor.mean(),group3_virginica.mean()
#mean of sepial width of all 3 category
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.scatter(range(1,51),group1_setosa)
plt.scatter(range(1,51),group2_versicolor)
plt.scatter(range(1,51),group3_virginica)

plt.legend(iris.target_names)
plt.show()
```

```python
#main part of command for anova a
from scipy import stats
stats.f_oneway(group1_setosa,group2_versicolor,group3_virginica)
#p value is 4.49 *10 ^-17 which is approximately equal to 0. so p<alpha
#so reject null hypo .so interpretation :at least 1 of them is different i.e setosa
```

```python
#another example of one way ANOVA
df1=pd.DataFrame(iris.data) #here in the abv we have iris data in data so v r making use of it
stats.f_oneway(df1.iloc[:,0],df1.iloc[:,1],df1.iloc[:,2],df1.iloc[:,3] ) #sl ,sw, pl ,pw
#p value is 4.66 *10 ^-17 which is approximately equal to 0.so reject null hypo
#H0=MUs1 = MUsW = MUP1 = MUPw , Ha= at least one of them is different
```

# **chi-square test**

```python
from scipy.stats import chi2_contingency
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#athlete and smoker example :these are observed frequencies
data=[[14,4],
     [0,10]]
data
```

```python
df=pd.DataFrame(data,
                index=['Athlete','Non Athlete'],
                columns=['Non smoker','smoker'])
df
```

```python
df.plot(kind='bar',
        stacked=True)
```

```python
chi2_contingency(data)#test stats, p value , df , expected frquency
#p 0.00038 < 0.005 so reject null hypothesis (both var are independent)
```

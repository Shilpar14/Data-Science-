```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
glaxo_df=pd.read_csv('/content/glaxo_df.csv')
glaxo_df
```

```python
beml_df=pd.read_csv('/content/beml_df.csv')
beml_df
```

```python
#here stats.norm.interval(0 will give u directly conf.interval for 95 % confidence .no need to  do another calculations
glaxo_df_ci=stats.norm.interval(0.95,
                                loc=glaxo_df.gain.mean(),
                                scale=glaxo_df.gain.std())
print("Mean at 95 % conf.interval iss :",np.round(glaxo_df_ci,4))
```

```python
beml_df_ci=stats.norm.interval(0.95,
                                loc=beml_df.gain.mean(),
                                scale=beml_df.gain.std())
print("Mean at 95 % conf.interval iss :",np.round(beml_df_ci,4))
```

```python
stats.t.ppf(0.975,139) #(percentage,degree of freedom) t value for 95 % confidence
#stats.t.ppf(0.95,139) #(percentage,degree of freedom) t value for 90 % confidence
```

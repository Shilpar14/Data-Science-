# **Isolation forest**

```python
#isolation forest example
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
df=pd.DataFrame({
    'Marks':[50,80,82,95,800],
    'Time':[1,2,3,1.5,60]
})
df
```

```python
#what is a outlier
iso=IsolationForest(random_state=10,contamination=0.1)#expecting 1% outlier in dataset
iso.fit(df)
out=iso.predict(df)
```

```python
out
```

```python
df['Score']=out
df
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
data=pd.read_csv("iris.csv",index_col=0)
data_encoded=pd.get_dummies(data)
```

```python
data_encoded
```

```python
# u can detect using precentile as well. other than this box plot can b used
data_encoded['Petal.Width'].describe(percentiles = [0.01, 0.02, 0.9,0.99,0.991,0.992])
```

```python
clf=IsolationForest(random_state=10,contamination=0.1)
clf.fit(data_encoded)
```

```python
#predictions
y_pred_outliers=clf.predict(data_encoded)
```

```python
data_encoded
```

```python
#training the model
clf=IsolationForest(random_state=10,contamination=0.1)
clf.fit(data_encoded)
y_pred_outliers=clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded['anamoly']=clf.predict(data_encoded.iloc[:,0:7])
```

```python
data_encoded
```

```python
#print the outliers
data_encoded[data_encoded['anamoly']==-1]#scores are given by decision tree
```

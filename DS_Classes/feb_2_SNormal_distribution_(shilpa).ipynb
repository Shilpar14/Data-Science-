```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
beml_df=pd.read_csv("BEML.csv")
beml_df.head()
```

```python
glaxo_df=pd.read_csv("GLAXO.csv")
glaxo_df.head()
```

```python
beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
```

```python
beml_df
```

```python
glaxo_df
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
#sort the beml df date
'''THe dataFrames have a date column ,so we can
create a datetimeIndex index from this column date.It will ensure that rows are sotred acoording to ascending order
'''
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
beml_df.head()#now date column is set as index
```

```python
glaxo_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close price')
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Time')
plt.ylabel('Close price')
```

```python
glaxo_df['gain']=glaxo_df.Close.pct_change(periods = 1) #pct_change does the work of formula for gain like (todays - yesterday)/yesterday  . and the period is like we are assuming only 1 day behind today so its 1 (yesterday)
beml_df['gain']=beml_df.Close.pct_change(periods = 1)
```

```python
glaxo_df
```

```python
beml_df
```

```python
#drop first row it is NAN
glaxo_df=glaxo_df.dropna()
beml_df=beml_df.dropna()
```

```python
glaxo_df
```

```python
beml_df
```

```python
#plot the gains
plt.figure(figsize=(8,5))
plt.plot(glaxo_df.gain)
plt.xlabel("Time")
plt.ylabel("gain")
```

```python
plt.figure(figsize=(8,5))
plt.plot(beml_df.gain)
plt.xlabel("Time")
plt.ylabel("gain")
```

```python
plt.figure(figsize=(10,5))
plt.suptitle('Distribution', fontsize=15)

plt.subplot(1,2,1)
sns.distplot(glaxo_df.gain,label='Glaxo')

plt.xlabel("GLAXO gain")
plt.ylabel("density")

plt.subplot(1,2,2)
sns.distplot(beml_df.gain,label='Beml')

plt.xlabel("BEML gain")
plt.ylabel("density")
```

```python
print('Mean:',round(glaxo_df.gain.mean(),4))# round function is we r rounding up to 4 values
print('Standard deviation:',round(glaxo_df.gain.std(),4))
```

```python
print('Mean:',round(beml_df.gain.mean(),4))
print('Standard deviation:',round(beml_df.gain.std(),4))
```
